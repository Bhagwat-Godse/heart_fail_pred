```python
df_heart.loc[df_heart['Sex']=="M",'Sex']=0
df_heart.loc[df_heart['Sex']=='F','Sex']=1

df_heart.loc[df_heart['ChestPainType']=='TA','ChestPainType']=0
df_heart.loc[df_heart['ChestPainType']=='ATA','ChestPainType']=1
df_heart.loc[df_heart['ChestPainType']=='NAP','ChestPainType']=2
df_heart.loc[df_heart['ChestPainType']=='ASY','ChestPainType']=3

df_heart.loc[df_heart['RestingECG']=='ST','RestingECG']=0
df_heart.loc[df_heart['RestingECG']=='Normal','RestingECG']=1
df_heart.loc[df_heart['RestingECG']=='LVH','RestingECG']=2

df_heart.loc[df_heart['ExerciseAngina']=='N','ExerciseAngina']=0
df_heart.loc[df_heart['ExerciseAngina']=='Y','ExerciseAngina']=1

df_heart.loc[df_heart['ST_Slope']=='Down','ST_Slope']=1
df_heart.loc[df_heart['ST_Slope']=='Flat','ST_Slope']=2
df_heart.loc[df_heart['ST_Slope']=='Up','ST_Slope']=3
```

```python
df_heart.info()
```

```python
df_heart['Sex']=df_heart['Sex'].astype(int)
df_heart['RestingECG']=df_heart['RestingECG'].astype(int)
df_heart['ST_Slope']=df_heart['ST_Slope'].astype(int)
df_heart['ChestPainType']=df_heart['ChestPainType'].astype(int)
df_heart['ExerciseAngina']=df_heart['ExerciseAngina'].astype(int)
```

```python
df_heart.info()
```

```python
df_heart.corr()['HeartDisease']
```

```python
X=df_heart.drop(['HeartDisease','RestingECG','RestingBP'],axis=1)
Y=df_heart['HeartDisease']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,random_state=40)
```

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBClassifier(colsample_bytree=0.43,learning_rate=0.0005,max_depth=8,n_estimators=250)
```

```python
xgb_model.fit(X_train,y_train)
pred1=xgb_model.predict(X_test)
pred1
```

```python
accuracy_score(pred1,y_test)
```

```python
def predict_class(Age, Sex, ChestPainType, Cholesterol, FastingBS, MaxHR, ExerciseAngina, Oldpeak, ST_Slope):
    data = np.array([[Age, Sex, ChestPainType, Cholesterol, FastingBS,MaxHR, ExerciseAngina, Oldpeak, ST_Slope]])
    prediction = xgb_model.predict(data)
    print("Predicted class is :",prediction[0])
    return prediction[0]
```

```python
predict_class(59,1,3,204,0,170,1,1.5,2)
```
